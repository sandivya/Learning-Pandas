```python
recent_grads['Major_category'].value_counts().sort_values(ascending = False)
```

```python
majorls = recent_grads['Major_category'].unique()
majorls
```

```python
major_mean = {}
for major in majorls:
    extract = recent_grads.loc[recent_grads['Major_category'] == major,:] 
    major_mean[major] = extract['Median'].mean()
    
major_mean
```

```python
plt.scatter(major_mean.keys(), major_mean.values())
plt.xticks(rotation = 90)
```
