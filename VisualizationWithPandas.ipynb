```python
# Import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Displaying plots in Jupyter
%matplotlib inline
```

```python
# increase default figure and font sizes for easier viewing
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

```python
# Reading data
drink_cols = ['country', 'beer', 'spirit', 'wine', 'liters', 'continent']
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(url, header = 0, names = drink_cols, na_filter = False)
```

```python
# Print first 5 rows of data
drinks.head()
```

# Histogram - Distribution of numerical variables

```python
#Extracted beer column values, sorted it and printed them
drinks.beer.sort_values().values
```

```python
# Divide these values in 3 bins(or groups, like 1 to 33 forms I group...)
drinks.beer.plot(kind = 'hist', bins =3)
```

```python
# Creating more bins, to view more granuarly
drinks.beer.plot(kind = 'hist', bins = 20)
```

```python
# Adding lables and titles to graph
drinks.beer.plot(kind = 'hist', bins = 10, title = 'Histogram of Beer Servings')
plt.xlabel('Beer Servings')
plt.ylabel('Count')
```

```python
# Density plot is the more smoother and outlined version of histogram
drinks.beer.plot(kind = 'density', xlim=(0,600))
```

# Bar Plot : compares different categories

```python
# Number of countries in each continent
drinks.continent.value_counts()
```

```python
drinks.continent.value_counts().plot(kind = 'bar')
```

```python
drinks.groupby('continent').mean()
```

```python
#Side-by-side bar plot
drinks.groupby('continent').mean().plot(kind = 'bar')
```

```python
#Above graph also contains a litre column which we don't want
drinks.groupby('continent').mean().drop('liters', axis = 1).plot(kind = 'bar')
```

```python
#stacking bar plot using stacked property
drinks.groupby('continent').mean().drop('liters', axis = 1).plot(kind = 'bar', stacked = True)
```

# Box Plot : show quartiles or outliers


**Five-number summary:**

min = minimum value
25% = first quartile (Q1) = median of the lower half of the data
50% = second quartile (Q2) = median of the data
75% = third quartile (Q3) = median of the upper half of the data
max = maximum value
(More useful than mean and standard deviation for describing skewed distributions)

Interquartile Range (IQR) = Q3 - Q1

Outliers:

below Q1 - 1.5 * IQR
above Q3 + 1.5 * IQR
